```python
# Fetch a webpage, check if the request was successful, and extract the main heading text from the page's content.
# You can assume that the base url is 'https://example.com' 

import requests
from bs4 import BeautifulSoup

url = 'https://example.com'
response = requests.get(url)

if response.status_code == 200:
    print("Successfully!")
    soup = BeautifulSoup(response.text, 'html.parser')
    main_heading = soup.find('h1') 
    if main_heading:
        print("The main heading is:", main_heading.text)
    else:
        print("Not found main heading.")
else:
    print("Failed to retrieve the page. Status code:", response.status_code)
```

```python
import requests

url = 'https://api.open-meteo.com/v1/forecast'
params = {
    'latitude': 52.52,
    'longitude': 13.41,
    'current': 'temperature_2m,wind_speed_10m',
    'hourly': 'temperature_2m,relative_humidity_2m,wind_speed_10m'
}

response = requests.get(url, params=params) 
content_type = response.headers.get('Content-Type', '')

print("Content-Type:", content_type)

if response.status_code == 200:
    print("Weather data retrieved successfully!")
    data = response.json()
    print(url)
    print(data)
   
else:
    print("Failed to retrieve weather data. Status code:", response.status_code)
```

```python
import requests

def fetch_weather():
    url = 'https://api.open-meteo.com/v1/forecast'
    
    params = {
        'latitude': 6.0329,
        'longitude': 80.2168,
        'current': 'temperature_2m,wind_speed_10m',
        'timezone': 'Asia/Colombo'
    }

    response = requests.get(url, params=params)

    if response.status_code == 200:
        content_type = response.headers.get('Content-Type', '')
        print("Content-Type:", content_type)
        print("Weather data retrieved successfully!")

        weather_data = response.json()
        print(weather_data)
    else:
        print("Failed to retrieve weather data. Status code:", response.status_code)
        return None

def parsing_weather_data(weather_data):
    if weather_data:
        if weather_data and 'current' not in weather_data:
            print("No current weather data available.")
            return None
        else:
            current_weather = weather_data.get('current', {})
            temperature = current_weather.get('temperature_2m')
            wind_speed = current_weather.get('wind_speed_10m')
        
        print(f"Current Temperature: {temperature}°C")
        print(f"Current Wind Speed: {wind_speed} m/s")
    else:
        print("No weather data to parse.")
        return None

fetch_weather()
parsing_weather_data(weather_data)
```
